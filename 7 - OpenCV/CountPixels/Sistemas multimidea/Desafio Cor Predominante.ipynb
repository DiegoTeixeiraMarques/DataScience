```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def Imagens_cor_predominante(x):
    #IMAGEM
    str(x)
    image = cv2.imread(x)
    #LISTA DE CORES
    cores = []
    #VARIAVEIS DAS CORES
    B = 0
    R = 0
    G = 0

    #PLOTANDO IMAGEM
    plt.imshow(image)
    plt.show()
    print(image.shape)

    #TAMANHO DA IMAGEM
    Eixox = image.shape[0]
    Eixoy = image.shape[1]
    print("Eixo X : "+str(Eixox) +" Eixo Y : "+str(Eixoy))

    #as cores são organizadas como B-blue G-green R-red respectivamente
    #print (image.item(100,400,0))#B
    #print (image.item(100,400,1))#G
    #print (image.item(100,400,2))#R

    #CONTABILIZANDO CORES B G R
    hist = cv2.calcHist([image], [1], None, [180], [0, 180])

    #variaveis bgr
    blue=0
    green=1
    red=2
    for x in range(0,image.shape[0]):
        for y in range(0,image.shape[1]):
            #if blue == 0 :
                B = B + image.item(x,y,blue)
           # elif green == 1:
                image.item(x,y,green)
                G = G + image.item(x,y,green)
            #elif red == 2:
                image.item(x,y,red)
                R = R + image.item(x,y,red)

    b = (B)
    g = (G)
    r = (R)
    
    cor_predominante = ' '

    cores.append(b)
    cores.append(g)
    cores.append(r)
    cores.sort(reverse=True)
    
    if cores[0] == b:
        cor_predominante = "cor predominantes na imagem é azul com " +str(b)+" pixels "+ str((b/cores[0]+cores[1]+cores[2])*100)+"%"
    elif cores[0] == g:
        cor_predominante = "cor predominantes na imagem é verde com " +str(g)+" pixels "+ str((g/cores[0]+cores[1]+cores[2])*100)+"%"
    elif cores[0] == r:
        cor_predominante = "cor predominantes na imagem é vermelho com " +str(r)+" pixels "+ str((r/cores[0]+cores[1]+cores[2])*100)+"%"
    
    return (cor_predominante)   
        #print(str(x) +" - "+ str(y))
    #for y in image:
    #    print (image.item(image[x],image[y],0))
```

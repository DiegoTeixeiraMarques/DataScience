A partir das 3 imagens existentes na pasta imagens no Github, crie os seguintes componentes em cada imagem e a salve posteriomente modificando o nome para o final “_processed.png”.

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.image as mpimg
```

**Imagem 1: rostos**

```python
image_rostos = cv2.imread("imagens/rostos.jpg")
cv2.destroyAllWindows()
cv2.waitKey()
#cv2.imshow("teste",image_rostos)
plt.imshow(image_rostos)
plt.show()
```

```python
###### Escreve seu código aqui
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.circle(image_rostos,(290,120), 55, (0,0,0), 0)
cv2.circle(image_rostos,(410,115), 60, (0,0,0), 0)
cv2.putText(image_rostos,'Mulher',(250,50), font, 1,(255,255,255),2,cv2.LINE_AA)
cv2.putText(image_rostos,'Mulher',(400,200), font, 1,(255,255,255),2,cv2.LINE_AA)

######
cv2.imwrite("rostos_processed.png", image_rostos)
image_plt = mpimg.imread("rostos_processed.png")
plt.imshow(image_plt)
plt.show()
```

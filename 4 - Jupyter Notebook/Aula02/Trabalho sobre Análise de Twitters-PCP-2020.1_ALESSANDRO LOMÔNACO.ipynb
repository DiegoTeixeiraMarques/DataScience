```python
# Selecionar colunas para an√°lise: create_at e full_text

# Bolsonaro
df_jb=pd.merge(df_jb.created_at.reset_index(),df_jb.full_text.reset_index())
df_jb.drop(columns=['index'],inplace=True)

# Lula
df_lu=pd.merge(df_lu.created_at.reset_index(),df_lu.full_text.reset_index())
df_lu.drop(columns=['index'],inplace=True)
```

```python
# Pegando arquivo com stopwords e colocando em uma lista. Todas as palavras em minusculo

arq_sw = pd.read_csv('stopwords.txt', header = None, names=['palavra'])
sw = []
for i in range(arq_sw.shape[0]):
    sw.append(str(arq_sw.palavra[i]).lower())
#sw
```

```python
# Seperando as palavras dos twitters atrav√©s do metodo word_tokenize

# Bolsonaro

palavras_jb = []
for i in range(df_jb.shape[0]):
    novo = nltk.word_tokenize(df_jb['full_text'][i])
    for j in range(len(novo)):
        palavras_jb.append(novo[j].lower())

# Lula

palavras_lu = []
for i in range(df_lu.shape[0]):
    novo = nltk.word_tokenize(df_lu['full_text'][i])
    for j in range(len(novo)):
        palavras_lu.append(str(novo[j]).lower())
```

```python
# Removendo StopWprds
palavras_jb_sem_sw = []
for palavra in palavras_jb:
    if palavra in sw:
        #print(palavra)
        pass
    else:
        palavras_jb_sem_sw.append(palavra)
```

```python
# Fazendo contagem de palavras

# Bolsonaro

dic_jb = dict()
for palavra in palavras_jb_sem_sw:
    if palavra in dic_jb:
        dic_jb[palavra] = dic_jb[palavra] + 1
    else:
        dic_jb[palavra] = 1
        
dic_jb = pd.DataFrame(dic_jb.items(), columns=['Palavra', 'QTD'])
dic_jb.sort_values(['QTD'],ascending=True)


for i in dic_jb.Palavra:
    print(i)
```

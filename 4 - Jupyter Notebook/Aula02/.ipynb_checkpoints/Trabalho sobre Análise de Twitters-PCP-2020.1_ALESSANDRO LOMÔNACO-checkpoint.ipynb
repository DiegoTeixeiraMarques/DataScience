```python
# Importando os M√≥dulos Pandas e NLTK

import pandas as pd
import nltk
#nltk.download()
```

```python
# Dados em DataFrame

df_jb = pd.read_json('jairbolsonaro.json')
df_lu = pd.read_json('LulaOficial.json')
```

```python
# Verificando o DataFrame do Bolsonaro com as 5 primeiras linhas

df_jb.head(5)
```

```python
# Verificando o DataFrame do Lula com as 5 primeiras linhas

df_lu.head(5)
```

```python
# Conhecendo os Dataframes

# Atribuindo informa√ß√µes: tweets, colunas e dados
qtdTweets_Lula = df_lu.shape[0]
qtdTweets_Bols = df_jb.shape[0]
dtFirst_Lula = df_lu.created_at.describe()[4].date()
dtFirst_Bols = df_jb.created_at.describe()[4].date()
dtLast_Lula = df_lu.created_at.describe()[5].date()
dtLast_Bols = df_jb.created_at.describe()[5].date()
qtdDays_Lula=(dtLast_Lula-dtFirst_Lula).days
qtdDays_Bols=(dtLast_Bols-dtFirst_Bols).days
mTweetsdia_Lula=round(qtdTweets_Lula/qtdDays_Lula,2)
mTweetsdia_Bols=round(qtdTweets_Bols/qtdDays_Bols,2)

# Criando um DataFrame com as informa√ß√µes
DFs = pd.DataFrame({'Qtd. de tweets': 
                              {'Bolsonaro': qtdTweets_Bols, 
                               'Lula': qtdTweets_Lula}, 
                          'Data 1¬∫ Tweet': 
                              {'Bolsonaro': dtFirst_Bols, 
                               'Lula': dtFirst_Lula},
                           'Data √öltimo Tweet': 
                              {'Bolsonaro': dtLast_Bols, 
                               'Lula': dtLast_Lula},
                           'Qtd.de dias no per√≠odo':
                              {'Bolsonaro':qtdDays_Bols,
                               'Lula':qtdDays_Lula},
                           'M√©dia Tweets por dia':
                              {'Bolsonaro':mTweetsdia_Bols,
                               'Lula':mTweetsdia_Lula}
                       })
DFs
```

```python
# Selecionar colunas para an√°lise 

df_jb=pd.merge(df_jb.created_at.reset_index(),df_jb.full_text.reset_index())
df_jb.drop(columns=['index'],inplace=True)
df_jb
```

```python
# Selecionar colunas para an√°lise 

df_lu=pd.merge(df_lu.created_at.reset_index(),df_lu.full_text.reset_index())
df_lu.drop(columns=['index'],inplace=True)
df_lu
```

```python
# Importanto stop words NLTK

sw = nltk.corpus.stopwords.words('portuguese')
# Add palavras ao stop words
#sw.append('')
#sw[len(sw)-1]
#sw = pd.DataFrame({'palavra': sw})


#arq_sw = pd.read_csv('stopwords.txt', header = None, names=['Palavra'])
#sw1 = arq_sw['Palavra'].str.lower().str.split()
#sw = []
#for i in range(sw1.shape[0]):
    #A = sw1[i]
    #print(A)
    #sw.append(A)
sw
```

```python
# Split dentro do full_text, ou seja, separando as palavras

palavras_jb = df_jb['full_text'].str.lower().str.split()
palavras_lu = df_lu['full_text'].str.lower().str.split()
palavras_lu
```

```python
# Removendo stopWords JB
palavras_jb_menos_sw = []
palavras_lu_menos_sw = []

for i in range(len(palavras_jb)):
    for j in range(len(palavras_jb[i])):
        if palavras_jb[i][j] not in sw:
            palavras_jb_menos_sw.append(palavras_jb[i][j])
            print("nao √© sw")
        else:
            print("achei")

for i in range(len(palavras_lu)):
    for j in range(len(palavras_lu[i])):
        if palavras_lu[i][j] not in sw:
            palavras_lu_menos_sw.append(palavras_lu[i][j])

#print(palavras_lu_menos_sw)
```

```python
# Conatabilizando as palavras
dictj = dict()

for Frases in palavras_jb_menos_sw:
    for Palavras in Frases:
            if Palavras in dictj:
                dictj[Palavras] = dictj[Palavras] + 1
            else: 
                dictj[Palavras] = 1
                
df_count_jb = pd.DataFrame(dictj.items(), columns=['Palavra', 'QTD'])
```

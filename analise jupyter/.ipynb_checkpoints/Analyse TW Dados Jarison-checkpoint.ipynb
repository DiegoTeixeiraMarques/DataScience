```python
Palavras_jair = dfj['full_text'].str.split()
```

```python
#criando dicion√°rio

dictj = dict()

for Frases in Palavras_jair:
    for Palavras in Frases:
            if Palavras in dictj:
                dictj[Palavras] = dictj[Palavras] + 1
            else: 
                dictj[Palavras] = 1

# print(dictj)
#for key in list(dictj.keys()): 
#    print(key, ":", dictj[key])
```

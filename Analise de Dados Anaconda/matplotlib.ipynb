```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(10,7))
ax.plot(x,y)

plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10, 7))
ax[0, 0].plot(x, y)
ax[0, 1].plot(x, y)
ax[0, 2].plot(x, y)
ax[1, 0].plot(x, y)
ax[1, 1].plot(x, y)
ax[1, 2].plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 10)
y = np.sin(x)

#plt.plot(x, y, "bo-") # Cada ponto do gráfico é um bola
plt.plot(x, y, "--") # Cada ponto do gráfico é um --
plt.plot(x, y, "x") # Cada ponto do gráfico é um x


plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Meu gráfico de linha")

plt.savefig("grafico.pdf")
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 10)
y = np.sin(x)

plt.plot(x, y, color='blue', linestyle='solid', linewidth=1, label='Sen x', marker='o')

y=np.cos(x)
plt.plot(x, y, color='#FF0000', linestyle='dashed', linewidth=4, label='Cos x')

plt.legend(fontsize=15, loc='upper right')
plt.xlabel('x', fontsize=30)
plt.ylabel('f(x)', fontsize=30)

plt.show()
```

```python
n=10000

x=np.random.rand(n)
y=np.random.normal(0,1,n)

plt.scatter(x,y)

plt.plot('x')
plt.plot('y')

plt.show()
```

```python
n=1000

x=np.random.rand(n)
y=np.random.rand(n)
cores=np.random.rand(n)

tamanhos=np.linspace(10,50,n)

mapa_de_cor=plt.cm.inferno # viridis, plasma, jet
plt.scatter(x,y, cmap=mapa_de_cor, c=cores, s=tamanhos, alpha=0.8)

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# Gráfico 3D sem renderização

from mpl_toolkits import mplot3d

fig=plt.figure()
ax=plt.axes(projection='3d')

z=np.linspace(0,15,1000)
x=np.sin(z)
y=np.cos(z)
ax.plot3D(x,y,z,'gray')

z=15*np.random.random(100)
x=np.sin(z)+0.1*np.random.randn(100)
y=np.cos(z)+0.1*np.random.randn(100)
p=ax.scatter3D(x,y,z,c=z,cmap='jet')

plt.show()
```

```python
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap='viridis')

plt.show()
```
